### Interseco de permisses

```python
drebin_list = list(drebin)
droid_list = list(droid)
krono_list = list(krono)
moto_list = list(moto)
```

```python
len(drebin_list)
```

```python
len(droid_list)
```

```python
len(krono_list)
```

```python
len(moto_list)
```

```python
extract = []
for x in drebin_list:
    if x in droid_list:
        if x in moto_list:
            if x in krono_list:
                extract.append(x)
```

```python
len(extract)
```

```python
extract
```

```python
drebin = drebin[extract]
droid = droid[extract]
krono = krono[extract]
moto = moto[extract]
```

### Exportando datasets

```python
drebin.to_csv("Intersection_Drebin.csv", index=False)
droid.to_csv("Intersection_DefenseDroid.csv", index=False)
krono.to_csv("Intersection_KronoDroid.csv", index=False)
moto.to_csv("Intersection_MotoDroid.csv", index=False)
```
