Testando se o DefenseDroid não possui muitos registros iguais aos do KronoDroid

```python
import pandas as pd
import numpy as np

# Tirando limitações de exibição de linhas e colunas
pd.options.display.max_rows = None
pd.options.display.max_columns = None
```

```python
droid = pd.read_csv('Intersection_DefenseDroid.csv')
krono = pd.read_csv('Intersection_KronoDroid.csv')
```

```python
droid.drop(columns=['Unnamed: 0'], inplace=True)
krono.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
print("DefenseDroid shape >>", droid.shape)
print("KronoDroid shape >>", krono.shape)
```

```python
df_concat = pd.concat([droid, krono])
df_concat.shape
```

```python
df_concat.head()
```

```python
print('Quantidade de exemplos duplicados entre os dois datasets >>',len(df_concat[df_concat.duplicated(keep='first')]))
```

```python
df_concat.drop_duplicates(keep="first", inplace=True)
```
